```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
tags = ['Query','HandleRange','HandleAggregation','HandleShuffle','HandleKeySwitch','Genesis','Insertion','CollectBitmap','GetBlock']
def dataToDF(path,vary):
    fp = open(path)
    tags.append(vary)
    final = []
    array = []
    for line in fp.readlines():
        if any(ext in line for ext in tags):
            toAppend = line.replace(" ","")
            toAppend = toAppend.replace('\n','')
            if(vary in toAppend):
                if(len(array)>1):
                    final.append(array)
                array = []
                value = toAppend.split(":")[1]
                if("." in value):
                    array.append(float(value))
                else:
                    array.append(int(value))
            else :
                value = toAppend.split(":")[1]
                array.append([float(s) for s in value.split(',')])
    final.append(array)
    fp.close()
    return final
```
