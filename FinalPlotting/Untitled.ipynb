```python
#Function to import stuff
import numpy as np
import matplotlib.lines as mlines
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import pylab
import re
import numpy as np
%matplotlib inline

def plot(path):
    data = []
    for p in path :
        fp = open(p)
        unit = "ms"

        for line in fp.readlines():
            if(line!=""):
                dataS = (re.findall(r'\d*[.]\d*',line))
                newUnit = (re.findall(r'[a-z|µ]{1,}',line))[0]
                if(newUnit[0]=='s'):
                    data.append(float(dataS[0])*1000)
                elif(newUnit[:2]=="µs"):
                    data.append(float(dataS[0])/1000)
                else :
                    data.append(float(dataS[0]))
    fig3, ax3 = plt.subplots()
    ax3.set_title("time in " + unit)
    print("mean is",np.mean(data))
    print("std is",np.std(data))
    ax3.boxplot(data)
    return np.mean(data),np.std(data)
```

```python
def autolabel(rects, ax):
    incremental = 0
    for rect in rects:
        y_val = rect.get_y() + rect.get_height() - rect.get_height() + 2
        if(rect.get_height()>1000):
            y_val += 150 + incremental
            incremental += 30
        if(rect.get_height()>12):
            y_val += 30
        ax.text(rect.get_x() + rect.get_width()/2., y_val,
                int(rect.get_height()),
                ha='center', va='center',fontsize=8)
```

```python
def unlynxTotal(p1,p2,p3,p4,p5):
    for r1,r2,r3,r4,r5 in zip(p1,p2,p3,p4,p5):
        h1 = r1.get_height()
        h2 = r2.get_height()
        h3 = r3.get_height()
        h4 = r4.get_height()
        h5 = r5.get_height()
        plt.text(r1.get_x()+r1.get_width()/2., h1+h2+h3+h4+h5, '%s'% int(h1+h2+h3+h4+h5), ha = 'center', va='bottom',fontsize=8)
```

```python
def prioTotal(p1,p2,p3):
    for r1,r2,r3 in zip(p1,p2,p3):
        h1 = r1.get_height()
        h2 = r2.get_height()
        h3 = r3.get_height()
        plt.text(r1.get_x()+r1.get_width()/2., h1+h2+h3, '%s'% int(h1+h2+h3), ha = 'center', va='bottom',fontsize=8)
```
