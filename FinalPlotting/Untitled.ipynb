```python
#Function to import stuff
import numpy as np
import matplotlib.pyplot as plt
import pylab
import re
import numpy as np
%matplotlib inline

def plot(path):
    data = []
    for p in path :
        fp = open(p)
        unit = "ms"

        for line in fp.readlines():
            if(line!=""):
                dataS = (re.findall(r'\d*[.]\d*',line))
                newUnit = (re.findall(r'[a-z|µ]{1,}',line))[0]
                if(newUnit[0]=='s'):
                    data.append(float(dataS[0])*1000)
                elif(newUnit[:2]=="µs"):
                    data.append(float(dataS[0])/1000)
                else :
                    data.append(float(dataS[0]))
    fig3, ax3 = plt.subplots()
    ax3.set_title("time in " + unit)
    print("mean is",np.mean(data))
    print("std is",np.std(data))
    ax3.boxplot(data)
    return np.mean(data),np.std(data)
```

```python
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),fontsize=4,
                ha='center')
```

```python
plt.figure(figsize=(4,4), dpi=200)
dpservRatio = np.ceil(np.array(y)/5)
xvall = np.multiply(dpservRatio,xv)
groups = [[xs,x,xp],[xi,xa,xdp,xsw,xv],[xi,xa,xdp,xsw,xvall]]
group_labels = y
num_items = len(group_labels)
# This needs to be a numpy range for xdata calculations
# to work.
ind = np.arange(num_items)

# Bar graphs expect a total width of "1.0" per group
# Thus, you should make the sum of the two margins
# plus the sum of the width for each entry equal 1.0.
# One way of doing that is shown below. You can make
# The margins smaller if they're still too big.
margin = 0.05
width = 0.3
#Colour for init
colour_classes = ['#ffcccc','#ff6666','#ff0000']
#Colour for aggregation
colour_agg = ['#adebad','#5cd65c',"#29a329"]
#Colour for Verif
colour_verif = ['#6666ff','#0000ff','#000099']
#Colour DP
colour_dp = ['#ffb84d','#e68a00']
#Colour switch
colour_sw = ['#d2a679','#ac7339']


s = plt.subplot(1,1,1)
for num, vals in enumerate(groups):
    # The position of the xdata must be calculated for each of the two data series
    xdata = ind+margin+(num*width)
    # Removing the "align=center" feature will left align graphs, which is what
    # this method of calculating positions assumes
    
    #PRIO
    if(num==0):
        p0 = plt.bar(xdata, vals[0], width, color=color_classes[num])
        p1 = plt.bar(xdata, vals[1], width,bottom=vals[0], color=colour_agg[num])
        p2 = plt.bar(xdata, vals[2], width,bottom=vals[1],color=colour_verif[num])
        autolabel(p0)
        autolabel(p1)
        autolabel(p2)
    plt.yscale('log')
    #Unlynx Anytrust
    if(num==1):
        p3 = plt.bar(xdata,vals[0],width, color=color_classes[num])
        p4 = plt.bar(xdata,vals[1], width,bottom=xi,color=colour_agg[num])
        p5 = plt.bar(xdata,vals[2],width,bottom=xa,color=colour_dp[num-1])
        p6 = plt.bar(xdata,vals[3],width,bottom=xdp,color=colour_sw[num-1])
        p7 = plt.bar(xdata,np.multiply(vals[4],y),width,bottom=xsw,color=colour_verif[num])
        autolabel(p3)
        autolabel(p4)
        autolabel(p5)
        autolabel(p6)
        autolabel(p7)
    #Unlynx All trusted
    if(num==2):
        p8 = plt.bar(xdata,vals[0],width, color=color_classes[num])
        p9 = plt.bar(xdata,vals[1], width,bottom=xi,color=colour_agg[num])
        p10 = plt.bar(xdata,vals[2],width,bottom=xa,color=colour_dp[num-1])
        p11 = plt.bar(xdata,vals[3],width,bottom=xdp,color=colour_sw[num-1])
        p12 = plt.bar(xdata,vals[4],width,bottom=xsw,color=colour_verif[num])
        autolabel(p8)
        autolabel(p9)
        autolabel(p10)
        autolabel(p11)
        autolabel(p12)
        plt.legend((p0[0],p1[0], p2[0],p3[0],p4[0], p5[0],p6[0],p7[0],p12[0]), ('Prio Initialization', 'Prio Aggregation','Prio input range validation','Unlynx Initialization', 'Unlynx Aggregation','Unlyn Data provider computation','Unlynx Key Switch','Unlynx AnyTrust Verfication','Unlynx all Trusted Verification'),bbox_to_anchor=(1., 1))

# You should no longer need to manually set the plot limit since everything 
# is scaled to one.
# Also the ticks should be much simpler now that each group of bars extends from
# 0.0 to 1.0, 1.0 to 2.0, and so forth and, thus, are centered at 0.5, 1.5, etc.
s.set_xticks(ind+0.5)
s.set_xticklabels(group_labels)
```
